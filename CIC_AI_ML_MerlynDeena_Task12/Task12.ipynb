##  Task 2: List Comprehension



```python



# 1. Square numbers
nums = [1, 2, 3, 4, 5]
squares = [x * x for x in nums]
print(squares)

```

```python


# 2. Find even numbers
nums = [1, 2, 3, 4, 5, 6]
even = [x for x in nums if x % 2 == 0]
print(even)


```

##  Task 3: Dictionary Comprehension





```python

# 1. Create number-square dictionary
nums = [1, 2, 3, 4]
squares = {x: x * x for x in nums}
print(squares)

```

```python
# 2. Filter even numbers
nums = [1, 2, 3, 4, 5, 6]
even_dict = {x: x * x for x in nums if x % 2 == 0}
print(even_dict)

```

```python
# 3. Map names to lengths
names = ["ai", "ml", "python"]
name_length = {name: len(name) for name in names}
print(name_length)
```

## Task 4: Set Practice

```python
# 1. Remove duplicates
nums = [1, 2, 2, 3, 4, 4, 5]
unique = set(nums)
print(unique)

```

```python
# 2. Find common elements (intersection)
a = {1, 2, 3}
b = {2, 3, 4}
common = a & b
print(common)

```

```python
# 3. Find union and difference
a = {1, 2, 3}
b = {3, 4, 5}

union = a | b
difference = a - b

print("Union:", union)
print("Difference:", difference)
```

##  Task 5: Generator Practice

```python
# 1. Create generator for numbers
def numbers(n):
    for i in range(n):
        yield i

for num in numbers(5):
    print(num)
```

```python
# 2. Generate even numbers
def even_numbers(n):
    for i in range(n):
        if i % 2 == 0:
            yield i

for num in even_numbers(10):
    print(num)
```

```python
# 3. Fibonacci using generator
def fibonacci(n):
    a, b = 0, 1
    for _ in range(n):
        yield a
        a, b = b, a + b

for num in fibonacci(7):
    print(num)
```

##  Task 6: Zip & Enumerate

```python
# 1. Combine 2 lists using zip
names = ["A", "B", "C"]
marks = [80, 90, 85]

combined = list(zip(names, marks))
print(combined)
```

```python
# 2. Print index + value using enumerate
names = ["AI", "ML", "Python"]

for i, name in enumerate(names):
    print(i, name)
```

```python
# 3. Convert zip result to dictionary
names = ["A", "B", "C"]
marks = [80, 90, 85]

result_dict = dict(zip(names, marks))
print(result_dict)
```
