# Task 2: Inheritance Practice
 1. Create a class Animal
 2. Create a class Dog inheriting Animal
 3. Add a method sound()

```python
class Animal:
    def sound(self):
        print("Animal makes sound")

class Dog(Animal):
    pass

d = Dog()
d.sound()
```

##  Task 3: Method Overriding

1. Parent class → Vehicle  
2. Child class → Car  
3. Override method start()

```python
class Vehicle:
    def start(self):
        print("Vehicle starts")

class Car(Vehicle):
    def start(self):
        print("Car starts with key")

c = Car()
c.start()
```

##  Task 4: Encapsulation Practice

1. Create class BankAccount  
2. Make balance private  
3. Create methods:  
   - deposit  
   - withdraw  
   - check balance  



```python
class BankAccount:
    def __init__(self, balance):
        self.__balance = balance   # private variable

    def deposit(self, amount):
        self.__balance += amount

    def withdraw(self, amount):
        self.__balance -= amount

    def check_balance(self):
        return self.__balance


# Example usage
acc = BankAccount(1000)
acc.deposit(500)
acc.withdraw(200)
print(acc.check_balance())
```

## Task 5: Combined Practice

Create:
- Class Person  
- Class Employee (inherits Person)  

Add:
- name  
- salary  
- display details

```python
class Person:
    def __init__(self, name):
        self.name = name

class Employee(Person):
    def __init__(self, name, salary):
        super().__init__(name)
        self.salary = salary

    def display(self):
        print("Name:", self.name)
        print("Salary:", self.salary)


# Example usage
e = Employee("Merlyn", 30000)
e.display()
```
