#  Task 3: Optimization Practice

1. Replace loops with built-in functions
2. Optimize sorting
3. Reduce repeated code

```python
#Replace loops with built-in functions
#Before
numbers = [1, 2, 3, 4, 5]
total = 0

for num in numbers:
    total += num

print(total)
```

```python
#After
numbers = [1, 2, 3, 4, 5]
total = sum(numbers)

print(total)
```

```python
# Optimize sorting
#Before
arr = [3, 1, 2]

for i in range(len(arr)):
    for j in range(i + 1, len(arr)):
        if arr[i] > arr[j]:
            arr[i], arr[j] = arr[j], arr[i]

print(arr)
```

```python
#After
arr = [3, 1, 2]
arr.sort()
print(arr)
```

```python
#Reduce repeated code
#Before
print(2 * 2)
print(4 * 4)
print(6 * 6)
```

```python
#After
def square(n):
    return n * n

print(square(2))
print(square(4))
print(square(6))
```

# Task 4: Refactoring
Convert: 
1. Long code → functions
2. Repeated logic → reusable function

```python
# 1. Long code → functions
# Before
numbers = [1, 2, 3, 4]

total = 0
for num in numbers:
    total += num

average = total / len(numbers)

print("Total:", total)
print("Average:", average)
```

```python
#After
def calculate_total(nums):
    return sum(nums)

def calculate_average(nums):
    return sum(nums) / len(nums)

numbers = [1, 2, 3, 4]

total = calculate_total(numbers)
average = calculate_average(numbers)

print("Total:", total)
print("Average:", average)
```

```python

#Repeated logic--> Reusable function
# Before
num = 5
if num > 0:
    print("Positive")
else:
    print("Negative")

num = -3
if num > 0:
    print("Positive")
else:
    print("Negative")

num = 10
if num > 0:
    print("Positive")
else:
    print("Negative")
```

```python
#After
def check_number(num):
    if num > 0:
        print("Positive")
    else:
        print("Negative")

check_number(5)
check_number(-3)
check_number(10)
```
