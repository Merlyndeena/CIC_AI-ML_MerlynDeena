## Task 4: Method Practice

### Create methods to:
- Check Pass/Fail
- Calculate Grade

```python
class Student:
    def __init__(self, name, marks):
        self.name = name
        self.marks = marks

    # Method to check pass or fail
    def check_result(self):
        if self.marks >= 40:
            return "Pass"
        else:
            return "Fail"

    # Method to calculate grade
    def calculate_grade(self):
        if self.marks >= 90:
            return "A+"
        elif self.marks >= 75:
            return "A"
        elif self.marks >= 60:
            return "B"
        elif self.marks >= 50:
            return "C"
        elif self.marks >= 40:
            return "D"
        else:
            return "F"

    # Method to display details
    def display(self):
        print("Name:", self.name)
        print("Marks:", self.marks)
        print("Result:", self.check_result())
        print("Grade:", self.calculate_grade())


# Creating object
s1 = Student("Merlyn", 78)

# Calling method
s1.display()
```

## Task 5: Object List

1. Create list of students  
2. Print all students  
3. Find topper

```python
class Student:
    def __init__(self, name, marks):
        self.name = name
        self.marks = marks

    def display(self):
        print("Name:", self.name, "| Marks:", self.marks)


# Create list of student objects
students = [
    Student("Merlyn", 78),
    Student("John", 85),
    Student("Alice", 92),
    Student("David", 67)
]

# Print all students
print("All Students:")
for s in students:
    s.display()

# Find topper
topper = students[0]
for s in students:
    if s.marks > topper.marks:
        topper = s

print("\nTopper:")
topper.display()
```

## Task 7: Logic Challenge

### 1. BankAccount Class
Methods:
- deposit
- withdraw
- check balance

### 2. Product Class
- name, price
- calculate discountc

```python
class BankAccount:
    def __init__(self, balance=0):
        self.balance = balance

    def deposit(self, amount):
        self.balance += amount
        print("Deposited:", amount)

    def withdraw(self, amount):
        if amount <= self.balance:
            self.balance -= amount
            print("Withdrawn:", amount)
        else:
            print("Insufficient balance")

    def check_balance(self):
        print("Current Balance:", self.balance)


# Product Class
class Product:
    def __init__(self, name, price):
        self.name = name
        self.price = price

    def calculate_discount(self, percent):
        discount = self.price * percent / 100
        final_price = self.price - discount
        return final_price


# Testing BankAccount
acc = BankAccount(1000)
acc.deposit(500)
acc.withdraw(300)
acc.check_balance()

# Testing Product
p1 = Product("Laptop", 50000)
print("Final Price after discount:", p1.calculate_discount(10))
```
