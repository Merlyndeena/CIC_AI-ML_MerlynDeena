# Task 2: JSON Practice



```python
# 1.Convert dictionary to JSON
import json
data = {"name": "Sara", "marks": 90}
json_data = json.dumps(data)
print(json_data)
```

```python

# 2. Convert JSON → Dictionary
import json
json_data = '{"name": "Sara", "marks": 90}'
data = json.loads(json_data)
print(data)
print(data["name"])
```

```python
# 3.Write JSON to file
import json
data = {"name": "Alex", "marks": 85}
with open("data.json", "w") as file:
    json.dump(data, file)
print("Data written to file")
```

```python
# 4. Read JSON from file
import json
with open("data.json", "r") as file:
    data = json.load(file)
print(data)
```

# Task 3: CSV Practice

```python
# 1 & 2. Create CSV File + Write Student Data
import csv

with open("students.csv", "w", newline="") as file:
    writer = csv.writer(file)
    
    # Header
    writer.writerow(["Name", "Marks"])
    
    # Student data
    writer.writerow(["John", 80])
    writer.writerow(["Sara", 90])
    writer.writerow(["Alex", 85])

print("CSV file created and data written!")
```

```python
# 3. Read CSV File
import csv

with open("students.csv", "r") as file:
    reader = csv.reader(file)
    
    for row in reader:
        print(row)
```

```python
# 4. Print Formatted Output
import csv
with open("students.csv", "r") as file:
    reader = csv.reader(file)  
    next(reader)  
    
    for row in reader:
        name = row[0]
        marks = row[1]
        
        print(f"Student: {name}, Marks: {marks}")
```

# Task 4: Combined Task
Program:
1. Read CSV file
2. Convert to dictionary
3. Save as JSON 

```python
import csv
import json

data_list = []

# Step 1: Read CSV and convert to dictionary
with open("students.csv", "r") as file:
    reader = csv.DictReader(file)
    
    for row in reader:
        data_list.append(row)

# Step 2: Save as JSON
with open("students.json", "w") as file:
    json.dump(data_list, file, indent=4)

print("CSV converted to JSON successfully!")
```
