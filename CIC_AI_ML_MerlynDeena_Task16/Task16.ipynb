Task 2: Cleaning Practice


```python
# 1. Remove None values 
students = [
    {"name": "A", "marks": 80},
    {"name": "B", "marks": None},
    {"name": "C", "marks": 40},
    {"name": "D", "marks": -10}
]

cleaned_students = []  

for s in students:
    if s["marks"] is not None and s["marks"] >= 0:
        cleaned_students.append(s)  

print("Cleaned Data:", cleaned_students)
```

```python
# 2. Remove Negative Values
students = [
    {"name": "A", "marks": 80},
    {"name": "B", "marks": None},
    {"name": "C", "marks": 40},
    {"name": "D", "marks": -10}
]

non_negative_students = []

for s in students:
    # keep marks that are NOT negative
    if s["marks"] is None or s["marks"] >= 0:
        non_negative_students.append(s)

print("After Removing Negative Values:", non_negative_students)
```

```python
# 3. Keep only valid data
students = [
    {"name": "A", "marks": 80},
    {"name": "B", "marks": None},
    {"name": "C", "marks": 40},
    {"name": "D", "marks": -10}
]

valid_students = []  # list to store valid data

for s in students:
    # check both conditions
    if s["marks"] is not None and s["marks"] >= 0:
        valid_students.append(s)

print("Valid Data:", valid_students)
```

Task 4: Sorting


```python
students = [
    {'name': 'A', 'marks': 80},
    {'name': 'B', 'marks': 60},
    {'name': 'C', 'marks': 40},
    {'name': 'D', 'marks': 90},
    {'name': 'E', 'marks': 70}
]

# 1. Sort students by marks (descending order)
students.sort(key=lambda x: x["marks"], reverse=True)

print("Sorted Students:", students)

# 2. Get Top 3 students
top_3 = students[:3]

print("Top 3 Students:", top_3)
```

Task 5: Grouping

```python
students = [
    {'name': 'A', 'marks': 80},
    {'name': 'B', 'marks': 60},
    {'name': 'C', 'marks': 40},
    {'name': 'D', 'marks': 90},
    {'name': 'E', 'marks': 30}
]

# Step 1: Create groups
grouped = {"pass": [], "fail": []}

for s in students:
    if s["marks"] >= 50:
        grouped["pass"].append(s)
    else:
        grouped["fail"].append(s)

print("Grouped Data:", grouped)

# Step 2: Count each group
print("Pass Count:", len(grouped["pass"]))
print("Fail Count:", len(grouped["fail"]))
```

Task 6: Data Analysis

```python
students = [
    {'name': 'A', 'marks': 80},
    {'name': 'B', 'marks': 60},
    {'name': 'C', 'marks': 40},
    {'name': 'D', 'marks': 90},
    {'name': 'E', 'marks': 30}
]

# Extract marks into a list
marks = [s["marks"] for s in students]

# 1. Average
average = sum(marks) / len(marks)

# 2. Highest
highest = max(marks)

# 3. Lowest
lowest = min(marks)

print("Average Marks:", average)
print("Highest Marks:", highest)
print("Lowest Marks:", lowest)
```
