# Task 2: Data Practice
1. Create list of marks
2. Find average
3. Find max and min
4. Count total students

```python
# Create list
marks = [80, 90, 70, 60, 50]

# Average
avg = sum(marks) / len(marks)
print("Average:", avg)

# Max & Min
print("Max:", max(marks))
print("Min:", min(marks))

# Count
print("Total Students:", len(marks))
```

# Task 3: Data Cleaning
1. Remove None values
2. Remove negative values 
3. Clean invalid data 


```python
marks = [80, 90, None, -10, 70, -5]

# Remove None
cleaned = [m for m in marks if m is not None]

# Remove negative values
cleaned = [m for m in cleaned if m >= 0]

print("Cleaned Data:", cleaned)
```

# Task 4: Data Filtering
1. Find passed students
2. Filter marks > 75
3. Filter even numbers 

```python
marks = [80, 40, 90, 30, 75, 60]

# Passed students
passed = [m for m in marks if m >= 50]

# Marks > 75
above_75 = [m for m in marks if m > 75]

# Even numbers
even = [m for m in marks if m % 2 == 0]

print("Passed:", passed)
print("Above 75:", above_75)
print("Even Marks:", even)
```

Task 5: Data Transformation
1. Convert marks to percentage
2. Multiply values
3. Normalize data

```python
marks = [80, 90, 70]

# Convert to percentage (0–1 scale)
percentage = [m / 100 for m in marks]

# Multiply values (e.g., scaling)
scaled = [m * 2 for m in marks]

# Normalize (simple max normalization)
max_val = max(marks)
normalized = [m / max_val for m in marks]

print("Percentage:", percentage)
print("Scaled:", scaled)
print("Normalized:", normalized)
```
