# Task 2: Arithmetic Practice

```python
import numpy as np

a = np.array([10, 20, 30])
b = np.array([2, 4, 5])

# 1. Add two arrays
print("Addition:", a + b)

# 2. Multiply arrays
print("Multiplication:", a * b)

# 3. Divide arrays
print("Division:", a / b)
```

# Task 3: Scalar Operations

```python
import numpy as np

arr = np.array([10, 20, 30])

# 1. Add constant to array
print("Add 5:", arr + 5)

# 2. Multiply array
print("Multiply by 2:", arr * 2)

# 3. Subtract constant
print("Subtract 3:", arr - 3)
```

# Task 4: Math Functions

```python
import numpy as np

arr = np.array([1, 4, 9, 16])

# 1. Find square root
print("Square Root:", np.sqrt(arr))

# 2. Find log values
print("Log Values:", np.log(arr))

# 3. Round values
values = np.array([1.2, 2.7, 3.5, 4.8])
print("Rounded:", np.round(values))
```

# Task 5: Aggregation

```python
import numpy as np

arr = np.array([10, 20, 30, 40, 50])

# 1. Find sum
print("Sum:", np.sum(arr))

# 2. Find mean
print("Mean:", np.mean(arr))

# 3. Find max and min
print("Maximum:", np.max(arr))
print("Minimum:", np.min(arr))
```

# Task 6: Axis Practice

```python
import numpy as np

arr = np.array([[10, 20],
                [30, 40]])

# 1. Row-wise sum
print("Row-wise Sum:", np.sum(arr, axis=1))

# 2. Column-wise sum
print("Column-wise Sum:", np.sum(arr, axis=0))
```

# Task 7: Filtering

```python

import numpy as np

arr = np.array([10, 15, 20, 25, 30, 35, 40])

# 1. Filter values > 20
print("Values > 20:", arr[arr > 20])

# 2. Filter even numbers
print("Even Numbers:", arr[arr % 2 == 0])
```
