# Task 2: Math Practice  

```python
import math

# 1. Find square root
print("Square Root:", math.sqrt(25))

# 2. Find power
print("Power:", math.pow(2, 3))

# 3. Round numbers
print("Round Up:", math.ceil(2.3))
print("Round Down:", math.floor(2.9))

# 4. Use π value
print("Pi Value:", math.pi)
```

# Task 3: Random Practice 

```python
import random

# 1. Generate random number
print("Random Number:", random.randint(1, 10))

# 2. Pick random element
numbers = [10, 20, 30, 40, 50]
print("Random Element:", random.choice(numbers))

# 3. Generate list of random numbers
random_list = []

for i in range(5):
    random_list.append(random.randint(1, 100))

print("Random List:", random_list)
```

# Task 4: Datetime Practice

```python
import datetime

# 1. Print current date
now = datetime.datetime.now()
print("Current Date:", now)

# 2. Format date
print("Formatted Date:", now.strftime("%Y-%m-%d"))

# 3. Extract year and month
print("Year:", now.year)
print("Month:", now.month)
```

# Task 5: Visualization

```python
import matplotlib.pyplot as plt

# Data
x = [1, 2, 3, 4, 5]
y = [10, 20, 15, 25, 30]

# 1. Create line graph
plt.plot(x, y)
plt.title("Line Graph")
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.show()

# 2. Create bar chart
plt.bar(x, y)
plt.title("Bar Chart")
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.show()

# 3. Plot student marks
students = ["John", "Sara", "Alex", "David"]
marks = [80, 90, 75, 85]

plt.bar(students, marks)
plt.title("Student Marks")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.show()
```
